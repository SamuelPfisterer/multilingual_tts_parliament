```python
import glob
import re

def collect_failed_urls():
    failed_urls = set()
    log_pattern = re.compile(r"Failed to process: (.*?)$|Error processing (.*?):")
    
    # Read all log files
    for log_file in glob.glob('logs/job_*_detailed.log'):
        print("logged")
        with open(log_file, 'r') as f:
            for line in f:
                if 'Failed to process' in line or 'Error processing' in line:
                    print("found failed")
                    match = log_pattern.search(line)
                    if match:
                        print("match")
                        url = match.group(1) or match.group(2)
                        url = "https://www.youtube.com/watch?v=" + url
                        if url:
                            failed_urls.add(url.strip())
    
    return list(failed_urls)

# Save failed URLs to a new CSV
def create_retry_csv():
    import pandas as pd
    
    failed_urls = collect_failed_urls()
    original_df = pd.read_csv('transcript_link_with_exactly_one_youtube_link.csv')
    
    # Filter original DataFrame for failed URLs
    retry_df = original_df[original_df['youtube_link'].isin(failed_urls)]
    
    # Save to new CSV
    retry_df.to_csv('failed_downloads.csv', index=False)
    print(f"Found {len(retry_df)} failed downloads")
    return retry_df
```

```python
retry_df = create_retry_csv()
```
